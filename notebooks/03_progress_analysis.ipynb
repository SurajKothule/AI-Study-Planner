```python
# 03_progress_analysis.ipynb

import pandas as pd
import json
import matplotlib.pyplot as plt

# Load user activity log
with open("../data/user_activity_log.json", "r") as f:
    data = json.load(f)

df_log = pd.DataFrame(data)

print("✅ User log:")
print(df_log)

# Daily study minutes
plt.figure(figsize=(10,6))
plt.plot(df_log["day"], df_log["minutes_studied"], marker="o")
plt.title("User Study Minutes Over Time")
plt.xlabel("Day")
plt.ylabel("Minutes Studied")
plt.grid(True)
plt.show()

# Completion status
completion_counts = df_log["completed"].value_counts()
plt.figure(figsize=(6,4))
completion_counts.plot(kind="bar", color=["green", "red"])
plt.title("Task Completion")
plt.xlabel("Completed")
plt.ylabel("Count")
plt.xticks(ticks=[0,1], labels=["True", "False"], rotation=0)
plt.show()

# Average minutes
avg_minutes = df_log["minutes_studied"].mean()
print(f"✅ Average daily study time: {avg_minutes:.1f} minutes")
```
