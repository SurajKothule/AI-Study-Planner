```python
# 01_EDA.ipynb

# 📌 Imports
%pip install matplotlib seaborn --quiet
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load Data
df = pd.read_csv("../data/raw_data.csv")

# Basic Info
print("✅ Data Info:")
print(df.info())
print("\n✅ Head:")
print(df.head())

# Summary Stats
print("\n✅ Describe:")
print(df.describe())

# Distribution plots
plt.figure(figsize=(12, 8))
sns.histplot(df["free_time_minutes"], bins=30, kde=True)
plt.title("Distribution of Free Time (Minutes)")
plt.xlabel("Free Time Minutes")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(12, 8))
sns.histplot(df["planned_minutes"], bins=30, kde=True, color="green")
plt.title("Distribution of Planned Minutes")
plt.xlabel("Planned Minutes")
plt.ylabel("Count")
plt.show()

# Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="free_time_minutes",
    y="planned_minutes",
    hue="engagement_score",
    size="days_completed",
    data=df,
    palette="viridis"
)
plt.title("Free Time vs Planned Minutes")
plt.xlabel("Free Time Minutes")
plt.ylabel("Planned Minutes")
plt.show()

# Correlation matrix
plt.figure(figsize=(8,6))
# Compute correlation only on numeric columns
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```
